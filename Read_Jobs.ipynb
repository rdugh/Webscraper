```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
folder='/Users/livi/Git/Webscraper/'
frame=[]
for i in os.listdir(folder):
    if i.endswith('.csv'):
        frame.append(pd.read_csv(i))

jobs=pd.concat(frame).drop_duplicates(subset='Description').reset_index(drop=True)
```

```python
jobs.head()
```

Cleaning Part

```python
for col in ['Post','Level','Function','Industry']:
    jobs[col]=jobs[col].str.lower()
```

```python
for col in ['Description']:
    jobs['Description']=jobs['Description'].apply(eval)
```

```python
#Post
sorted(Counter(jobs['Post']).items(), key=lambda x: x[1],reverse=True)
```

```python
for col in ['Level','Function','Industry']:
    rank=Counter(jobs[col])
    rank_sort=sorted(rank.items(), key=lambda x: x[1],reverse=True)
    
    plt.figure(figsize=(10,5))
    plt.bar([i[0] for i in rank_sort],[i[1] for i in rank_sort])
    plt.xticks(rotation=90)
    plt.show()
```
